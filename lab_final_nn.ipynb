```python
with open('/content/TGDataset.txt', 'r') as file:
   file_content = file.read()
```

```python
file_content
```

```python
import tensorflow as tf
from tensorflow.keras.preprocessing.text import Tokenizer
```

```python
tokenizer = Tokenizer()
```

```python
tokenizer.fit_on_texts([file_content])
```

```python
tokenizer.word_index
```

```python
len(tokenizer.word_index)
```

```python
input_sequences = []
for sentence in file_content.split('\n'):
  tokenized_sentence = tokenizer.texts_to_sequences([sentence])[0]

  for i in range(1,len(tokenized_sentence)):
    input_sequences.append(tokenized_sentence[:i+1])
```

```python
input_sequences
```

```python
max_len = max([len(x) for x in input_sequences])
```

```python
max_len
```

```python
from tensorflow.keras.preprocessing.sequence import pad_sequences
padded_input_sequences = pad_sequences(input_sequences, maxlen = max_len, padding='pre')
```

```python
padded_input_sequences
```

```python
X = padded_input_sequences[:,:-1]
```

```python
y = padded_input_sequences[:,-1]
```

```python
y
```

```python
X.shape
```

```python
from tensorflow.keras.utils import to_categorical
y = to_categorical(y,num_classes=4994)
```

```python
y.shape
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding, LSTM, Dense
```

```python
model = Sequential()
model.add(Embedding(4994, 100, input_length=325))
model.add(LSTM(150))
model.add(Dense(4994, activation='softmax'))
```

```python
model.compile(loss='categorical_crossentropy', optimizer='adam',metrics=['accuracy'])
```

```python
model.summary()
```

```python
model.fit(X,y,epochs=50)
```

```python
import numpy as np
import time
text = "As the day turned"

for i in range(15):
  # tokenize
  token_text = tokenizer.texts_to_sequences([text])[0]
  # padding
  padded_token_text = pad_sequences([token_text], maxlen=20, padding='pre')
  # predict
  pos = np.argmax(model.predict(padded_token_text))

  for word,index in tokenizer.word_index.items():
    if index == pos:
      text = text + " " + word
      print(text)
      time.sleep(0)
```
